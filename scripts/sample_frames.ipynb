```python
import cv2
```

```python
## We are going to sample frames using this tool

PATH_VIDEO = "/home/suyog/downloads/dataset.mp4" # Path to the 13 minute video

OUTPUT_PATH = "/home/suyog/desktop/projects/microlane/results/13-minute-sampling"
```

```python
video = cv2.VideoCapture(PATH_VIDEO)

fps = video.get(cv2.CAP_PROP_FPS)
```

```python
frame_interval = int(fps * 1.5) # 1 frame every 1.5 seconds

frame_count = 0
saved = 0

while video.isOpened():
    
    ret, frame = video.read()
    
    if not ret:
        break
    
    if frame_count % frame_interval ==0:
        cv2.imwrite(f"{OUTPUT_PATH}/frame_{saved:04d}.jpg", frame)
        saved+=1
    frame_count+=1

video.release()
```

```python
print(f"Frame Count: {frame_count}")
print(f"Total Saved: {saved}")
```
