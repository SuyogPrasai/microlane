```python
# Read the json
# Then resd the file line by line
# Each line is a json object
# modify th eict to have the same format as the original json file
# Then write the modified dict to a new json file
```

```python
import json
```

```python
PATH = "/home/suyog/assets/datasets/TuSimple/test_label_new.json"
OUTPUT_PATH = "/home/suyog/assets/datasets/TuSimple/test_label_path_normalised.json"
BASE = "/home/suyog/assets/datasets/TuSimple/TUSimple/"

```

```python
with open(PATH, "r") as f:
    lines = f.readlines()

with open(OUTPUT_PATH, "w") as out:
    for line in lines:
        line = line.strip()
        if not line:
            continue
        obj = json.loads(line)
        obj["raw_file"] = BASE + obj["raw_file"]
        out.write(json.dumps(obj) + "\n")
```
